```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset,DataLoader
from torchsummary import summary
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
device= 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
data_df=pd.read_csv('/content/rice-type-classification/riceClassification.csv')
data_df.head()
```

```python
data_df.dropna(inplace=True)
data_df.drop(['id'],axis=1,inplace=True)
print(data_df.shape)
```

```python
data_df.head()
```

```python
original_df=data_df.copy()

for column in data_df.columns:
  data_df[column]= data_df[column]/data_df[column].abs().max()

data_df.head()
```

```python
x= np.array(data_df.iloc[:, :-1])
y= np.array(data_df.iloc[:, -1])
print(x.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3)
```

```python
x_test,x_val,y_test,y_val= train_test_split(x_test,y_test,test_size=0.5)
```

```python
x_val.shape
```

```python
class dataset(Dataset):
  def __init__(self,x,y):
    self.x=torch.tensor(x,dtype=torch.float32).to(device)
    self.y=torch.tensor(y,dtype=torch.float32).to(device)

  def __len__(self):
    return len(self.x)

  def __getitem__(self, index):
    return self.x[index], self.y[index]
```

```python
train_data=dataset(x_train,y_train)
val_data=dataset(x_val,y_val)
test_data=dataset(x_test,y_test)
```

```python
train_dataloader=DataLoader(train_data,batch_size=32,shuffle=True)
val_dataloader=DataLoader(val_data,batch_size=32,shuffle=True)
test_dataloader=DataLoader(test_data,batch_size=32,shuffle=True)
```

```python
class model(nn.Module):
  def __init__(self):
    super(model,self).__init__()
    self.fc1=nn.Linear(x.shape[1],10)
    self.fc2=nn.Linear(10,1)
    self.sig=nn.Sigmoid()

  def forward(self,x):
    x=self.fc1(x)
    x=self.fc2(x)
    x=self.sig(x)
    return x

```

```python
model=model().to(device)
```

```python
summary(model,(x.shape[1],))
```

```python
import torch.optim as optim
criterion=nn.BCELoss()
optimizer=optim.Adam(model.parameters(),lr=0.001)
```

```python

total_loss_train_plot=[]
total_loss_val_plot=[]
total_acc_train_plot=[]
total_acc_val_plot=[]

```

```python
for epoch in range(10):
  total_loss_train=0
  total_loss_val=0
  total_acc_train=0
  total_acc_val=0

  for data in train_dataloader:
    x,y=data

    pred=model(x).squeeze(1)

    batch_loss=criterion(pred,y)

    total_loss_train+=batch_loss.item()

    acc=((pred).round()==y).sum().item()

    total_acc_train+=acc

    batch_loss.backward()
    optimizer.step()
    optimizer.zero_grad()

  with torch.no_grad():
    for data in val_dataloader:
      x,y=data

      pred=model(x).squeeze(1)

      batch_loss=criterion(pred,y)

      total_loss_val+=batch_loss.item()

      acc=((pred).round()==y).sum().item()

      total_acc_val+=acc
total_loss_train_plot.append(round(total_loss_train/1000,4))
total_loss_val_plot.append(round(total_loss_val/1000,4))

total_acc_train_plot.append(round(total_acc_train/train_data.__len__()*100,4))
total_acc_val_plot.append(round(total_acc_val/val_data.__len__()*100,4))
print(f'''epoch no: {epoch+1}
      TrainLoss:{round(total_loss_train/1000,4)}
      TrainAcc:{round(total_acc_train/train_data.__len__()*100,4)}
      ValiLoss:{round(total_loss_val/1000,4)}
      ValiAcc:{round(total_acc_val/val_data.__len__()*100,4)}''')
print('-----'*20)
```

```python
with torch.no_grad():
  total_loss_test=0
  total_acc_test=0
  for data in test_dataloader:
    x,y=data

    pred=model(x).squeeze(1)

    batch_loss_t=criterion(pred,y).item()
    total_loss_test+=batch_loss_t

    acc=((pred).round()==y).sum().item()
    total_acc_test+=acc
print('Acc: ',round(total_acc_test/test_data.__len__()*100,4))
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
ax[0].plot(total_loss_train_plot,label='training loss')
ax[0].plot(total_loss_val_plot,label='validation loss')
ax[0].set_title('train and val loss over epochs')
ax[0].set_xlabel('epochs')
ax[0].set_ylabel('loss')
ax[0].set_ylim([0,0.1])
ax[0].legend()

ax[1].plot(total_acc_train_plot,label='training acc')
ax[1].plot(total_acc_val_plot,label='validation acc')
ax[1].set_title('train and val acc over epochs')
ax[1].set_xlabel('epochs')
ax[1].set_ylabel('acc')
ax[1].set_ylim([90,100])
ax[1].legend()

plt.show()
```
